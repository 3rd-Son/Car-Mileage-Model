##### *Some visualizations and statistical analysis*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
preprocessed_data = pd.read_csv('/Users/appleplay/Desktop/Car Mileage Model/data/processed/preprocessed_data.csv')
```

```python
preprocessed_data.head()
```

##### *Corrrelation plots*

```python
attributes = preprocessed_data.iloc[:,0:7]
sns.pairplot(attributes,diag_kind='kde')
plt.show()
```

```python
preprocessed_data.plot(kind='box',subplots=True,layout=(3,3),sharey=False,figsize=(12,7),title='Box Plot')
plt.show()
```

```python
preprocessed_data['cylinders'].value_counts()
```

```python
preprocessed_data.plot(kind='density',subplots=True,layout=(3,3),sharex=False,figsize=(12,7),title='Kde Plot')
plt.show()
```

```python
sns.heatmap(preprocessed_data.corr(),annot=True)
plt.figure(figsize=(12,6))
plt.show()
```
